```python
# An object of Flask class is our WSGI application.
from flask import Flask

# Flask constructor takes the name of 
# current module (__name__) as argument.
app = Flask(__name__)

# The route() function of the Flask class is a decorator, 
# which tells the application which URL should call the associated function.
@app.route('/')
# ‘/’ URL is bound with hello_world() function.
def flaks_app_home():
    return "Welcome to Home page of Flask application"

# main driver function
if __name__ == '__main__':

    # run() method of Flask class runs the application on the local development server.
    app.run()
```

```python
# Defining home route of app

@app.route("/")
def flaks_app_home():
    return "Welcome to Home page of Flask application"
```

```python
from flask import Flask,request

add_app=Flask(__name__)

@add_app.route("/add",methods=["GET"])
# ‘/add’ URL is bound with add_two_numbers() function.
def add_two_numbers():

    """
    In general, request.args.get retrieves query parameters as strings by default, 
    because HTTP query parameters are typically passed as text.
    """
    num1 = request.args.get("num1", type=float)  # Default type is changed to float for flexibility
    num2 = request.args.get("num1", type=float)  # Default type is changed to float for flexibility

    if num1 is None or num2 is None:
        return "Please provide both 'num1' and 'num2' as query parameters.", 400
    
    return f"The sum of two {num1} and {num2} is {num1+num2}"

if __name__=="__main__":
    add_app.run(debug=True)
```
